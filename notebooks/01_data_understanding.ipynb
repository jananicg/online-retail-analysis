```python
df_sales["InvoiceDate"] = pd.to_datetime(
    df_sales["InvoiceDate"],
    errors="coerce"
)
```

```python
df_sales = df_sales.copy()
```

```python
df_sales["InvoiceDate"] = pd.to_datetime(
    df_sales["InvoiceDate"],
    errors="coerce"
)
```

```python
df_sales["InvoiceDate"].dtype
```

```python
df_sales["Month"] = df_sales["InvoiceDate"].dt.to_period("M")
```

```python
df_sales[["InvoiceDate", "Month"]].head()
```

```python
monthly_revenue = (
    df_sales
    .groupby("Month")["Revenue"]
    .sum()
    .reset_index()
)
```

```python
monthly_revenue.head()
```

```python
df_sales["Month"].nunique()
```

```python
monthly_revenue
```

```python
monthly_revenue["Month"].min(), monthly_revenue["Month"].max()
```

```python
monthly_revenue.tail()
```

```python
country_revenue = (
    df_sales
    .groupby("Country")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
country_revenue
```

```python
top_customers = (
    df_sales
    .groupby("CustomerID")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
top_customers
```

```python
top_customer_revenue = top_customers.iloc[0]
total_revenue = df_sales["Revenue"].sum()

(top_customer_revenue / total_revenue) * 100
```
